### Answer 1: 

```python
import numpy as np
import matplotlib.pyplot as plt

def least_mean_sqaure(x,y):    
    x_squared = [i**2 for i in x]
    xy = [a * b for a, b in zip(x, y)]
    x_sum = sum(x)
    y_sum = sum(y)
    x_squared_sum = sum(x_squared)
    xy_sum = sum(xy)

    m = ((len(x)*(xy_sum))-(x_sum*y_sum))/((len(x)*x_squared_sum)-((x_sum)**2))
    b = ((y_sum)-(m*x_sum))/(len(x))
    return m,b

q1_x = np.array([-2,1,3])
q1_y = np.array([-1,1,2])
que_1 = least_mean_sqaure(q1_x,q1_y)

print("Answer 1(a): The least square regression line for the given data points is y = ", que_1[0] , "+ x * ", que_1[1])

print("\nAnswer 1(b): ")
plt.scatter(q1_x,q1_y, c='r')
plt.plot(q1_x,que_1[0]*q1_x+que_1[1])
```
